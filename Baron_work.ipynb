"Does Genre correlate to Popularity?"

```python
import pandas as pd
import numpy as np
import json
import sys
import sklearn.datasets
import csv
import re
from functools import reduce
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#TestSDA.csv is the name of the file containing the dataset.
data = pd.read_csv('TestSDA.csv')  
```

```python
data.sort_values(by=['Popularity'], ascending=False)
```

```python
sns.catplot(data=data, kind="bar", x="Genre", y="Popularity", aspect=14, height=2)
```

From what this dataset shows there doesn’t seem to be a correlation between popularity and genre due to the most popular songs all being from different genres.
